```python
import requests
from bs4 import BeautifulSoup
import re
import csv
import json

# Initialize an empty list to store the data
data_list = []
paginated_set_urls=[]

def extract_number_or_none(url):
    parts = url.split('/')
    for part in parts:
        if part.startswith('p'):
            number_str = part[1:]
            try:
                return int(number_str)
            except ValueError:
                pass
    return -1

def function1(initial_url):
    href_set = set()
    urls_to_check = [initial_url]

    while urls_to_check:
        current_url = urls_to_check.pop(0)
        response = requests.get(current_url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")

            # Find all links and store their href values in the set
            for link in soup.find_all('a'):
                href = link.get('href')
                if href and href.startswith('https://community.breastcancer.org/en/categories/alternative-medicine'):
                    href_set.add(href)

            # Regular expression pattern to match pagination URLs
            pagination_pattern = r'https://community\.breastcancer\.org/en/categories/alternative-medicine/p\d+'
            pagination_regex = re.compile(pagination_pattern)

            # Find pagination URLs on the current page
            pagination_links = soup.find_all('a', href=pagination_regex)

            # Extract the page numbers from the pagination links
            page_numbers = [int(re.search(r'/p(\d+)', link.get('href')).group(1)) for link in pagination_links if re.search(r'/p(\d+)', link.get('href'))]

            # Append all paginated URLs within the range to the list
            if page_numbers:
                for page_number in range(min(page_numbers), max(page_numbers) + 1):
                    paginated_url = f"https://community.breastcancer.org/en/categories/alternative-medicine/p{page_number}"
                    if paginated_url not in href_set:
                        urls_to_check.append(paginated_url)
                        href_set.add(paginated_url)

    # Return the sorted list of URLs
    return sorted(href_set, key=extract_number_or_none)



def function2(urls):
    discussion_urls = []
    # Process each URL in the urls list
    for url in urls:
        # Send a GET request to the URL
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            discussion_pattern = r'https://community\.breastcancer\.org/en/discussion/\d+/[^/]+(?:/p\d+)?'
            discussion_regex = re.compile(discussion_pattern)
            # Filter discussion thread URLs
            discussion_urls.extend(filter(discussion_regex.match, [link.get('href') for link in soup.find_all('a')]))

    return discussion_urls

def function3(discussion_urls):
    paginated_urls = []
    for discussion_url in discussion_urls:
        response = requests.get(discussion_url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            # Find all pagination links on the page
            pagination_links = soup.find_all('a', href=re.compile(rf'{re.escape(discussion_url)}/p\d+'))
            if pagination_links:
                # Extract the page numbers from the pagination links
                page_numbers = [int(link.text) for link in pagination_links if link.text.isdigit()]
                # Append all paginated URLs within the range to the list
                for page_number in range(min(page_numbers), max(page_numbers) + 1):
                    paginated_urls.append(discussion_url)
                    paginated_urls.append(f"{discussion_url}/p{page_number}")

    return list(set(paginated_urls))

# Example usage
initial_url = "https://community.breastcancer.org/en/categories/alternative-medicine"
urls = function1(initial_url)
print("All URLs:", urls)


discussion_urls = function2(urls)
print("Discussion URLs:", discussion_urls)

paginated_urls = function3(discussion_urls)
print("Paginated URLs:", list(set(paginated_urls)))

paginated_set_urls = []
set_urls=[]
# paginated_set_urls.extend(urls)
paginated_set_urls.extend(discussion_urls)
paginated_set_urls.extend(paginated_urls)
# set_urls = sorted(paginated_set_urls, key=extract_number_or_none)
print("All URLs sorted by pagination:", paginated_set_urls)
```
